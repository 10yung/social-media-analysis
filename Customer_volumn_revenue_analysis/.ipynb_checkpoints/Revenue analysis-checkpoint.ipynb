### 1.1 Time Series line chart
Draw time series line chart. The vertical dashline means the difference of the status.

```python
def time_window_selector(df, start_date: str, end_date: str):
    # 2010-10-01
    start_date = datetime.strptime(start_date, '%Y-%m-%d')
    end_date = datetime.strptime(end_date, '%Y-%m-%d')
    interval = df.query('Date >= @start_date & Date <= @end_date')
    
    return interval

def columns_line_plotter(df, columns_name: list, mobile_launched_date = True, date_axis = True) -> None:
    fig=plt.figure(figsize=(16, 12))
    # Check if use column `date` as x-axis   
    if date_axis: 
        for column_name in columns_name:
            plt.plot('Date', column_name, data=df, marker='', linewidth=1)
        # Check is app aluncah date line is included
        if mobile_launched_date:
            plt.axvline(x=first_app_launched_date.to_pydatetime(), color='k', linestyle='dashed', linewidth=1)
    else:
        for column_name in columns_name:
            plt.plot(column_name, data=df, marker='', linewidth=1)
    
    plt.legend()
```

# Time series predictive analysis
Assume that the time series data we observe is from $f(x) = Signal + White\ Noise$. Here, $Signal$ is denoted as the pattern we want to model and $White\ Noise$ is the randomness that we cannot capture. Note: $White\ Noise$ is a series that with mean $\mu$ is $0$, constant $\sigma$, and correlation between lag is $0$. On the other hand, if we can prove that our residual is close to $White\ Noise$ then, we can say we have a good model that can predict our data.

1. Check the stationary and corelated of time series
    - Stationary means:
        - $\sigma$ is constant through time
        - No seasonality, no predictable pattern shows in the data
        - The mean $\mu$ of the data is constant
    - Ways to check stationary
        - Visually
        - Global V.S Local test - slice different interval to test the difference
        - ADF Test
2. See how to erase non-stationary effect or include variables as factors for time series pattern.
        
3. Choose model ( 2 ways )
    - Time series model
    - Add dummmy variable
    
4. Identify prediciton power of our model (good or bad?)
    - Check if the resdual is close to white noise
        - Ways to check white noise
            - Rolling window mean should equal to each other (constant $\sigma$)
            - Visually
            - Check ACF/PACF to see if there is no autocorrelation

### Augmented Dickeyâ€“Fuller (ADF)
Traditional statistical time series models rely on a time series being stationary. Which means that the observe data has fairly stable statistical properties over time ( No depend factors, called `unit root` , that effect the representation of the data ). ADF is a test to test is the observed data is stationary or not.

$Y_t = \phi \times Y_{t-1} + \epsilon_t $

If the $\phi$ equal to $1$, which means there $Y_t$ is depend on the previous data $Y_{t-1}$. Means there this observesd time series data is not stationary.

If you are seeking a mean from a non stationary series. There will be lots of bias and error

$H_0:$ Time series has a unit root, meaning it is non-stationary.

$H_a:$ Time series has no unit root, meaning it is stationary.

Given $\alpha$ is equal to 0.05


```python
# Given alpha is equal to 0.05
for column in numerical_columns:
    result = adfuller(df[column])
    if result[1] <= 0.05:
        stationary_or_not = 'Stationary'
    else:
        stationary_or_not = 'Non-stationary'
        
    print(f'The P-value of {column} is {result[0]} => {stationary_or_not}')
```

### Self-Correlation Checking
To check a value in a time series at one given point in time may have a correlation to the value at another point in time.

* Autocorrelation: It's correlation of a signal with a delayed copy of itself. Informally, it is the similarity between observations as a function of the time lag between them. 
* It generalizes self-correlation by not anchoring to a specific point in time.
*  Assume the distribution of each variable fits a Gaussian (bell curve) distribution. 


- Reference

    - https://dzone.com/articles/autocorrelation-in-time-series-data



```python
# The shallow blue cone area is the 95% confident interval that the correlation number fall in the zone
def acf_plotter(df, columns_name: list, lags):
    number_of_subplot = len(columns_name)
    fig, axs = plt.subplots(number_of_subplot, figsize=(16, 32))
    for i, column in enumerate(columns_name):
        plot_acf(df[column], lags=lags, ax=axs[i])
        axs[i].set_title(column)
    plt.show()
    
acf_plotter(df, numerical_columns, 100)
```

```python
# partila corelated diagram
def pacf_plotter(df, columns_name: list, lags):
    number_of_subplot = len(columns_name)
    fig, axs = plt.subplots(number_of_subplot, figsize=(16, 32))
    for i, column in enumerate(columns_name):
        plot_pacf(df[column], lags=lags, ax=axs[i])
        axs[i].set_title(column)
    plt.show()
    
pacf_plotter(df, numerical_columns, 100)
```

## 1. Autoregression (AR) model
Observed data $y_t$ is effected by the sum of previous value times a weight. The AR(q) is denoted as $y_t = bias + \sum_{n = 1}^{q}\phi_n \times y_{t-n} + \epsilon_t$. The param $q$ represent the lag of time, which means how many "pre-step" of data we should take into consideration while building our $AR(q)$ model.

Here is the example of $AR(1)$: 

$y_t=b_0+b_1Ã—y_{tâˆ’1}+\epsilon_t$

This model means we predict our $y_t$ base on one step ahead data with a error term and a bias.

### 1.1 The Search AD  prediction base on AR model
Predict $Search \thinspace AD_{7}$ by adopting $Search \thinspace AD_{1-t}$

```python
def plot_model_result(time_step_index, target_name: str, pred: list, test_data: list, residual: list) -> None:
    '''
    plot the result base on observed data (test data), prediction and residual
    
    return none
    '''
    # calcualte residual
    residual = ar_test - predictions

    prediction  = pd.DataFrame(data = pred)
    observed_data = pd.DataFrame(data = test_data)
    result = pd.concat([observed_data, prediction], axis = 1, ignore_index=False)
    result.index = time_step_index

    fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(16, 4))
    result.plot(ax=axes[0])

    axes[0].set_title('Observed Data and Prediction')
    axes[1].plot(residual)
    axes[1].set_title('Residual')
    plt.show()


# Separate training and testing set
ar_train_df, ar_test_df = df[1:len(df)-7], df[len(df)-7:]
ar_train = ar_train_df['Search_AD'].values
ar_test = ar_test_df['Search_AD'].values
test_index = ar_test_df.index

# build AR model
model = AR(ar_train)
model_fit = model.fit()

# Train AR model
predictions = model_fit.predict(start=len(ar_train), end=len(ar_train)+len(ar_test)-1, dynamic=False)

residual = ar_test - predictions
error = mean_squared_error(ar_test, predictions)
# for i in range(len(predictions)):
#     print('predicted=%f, expected=%f' % (predictions[i], ar_test[i]))
print('Lag: %s' % model_fit.k_ar)
print('Test MSE: %.3f' % error)
# print('Coefficients: %s' % model_fit.params)

plot_model_result(test_index, 'Search_AD', predictions, ar_test,residual)
```

## 2. Moving Average Model (MA)
Observed data is effected by the error $\epsilon$, the prediction and real value, of previous steps. The function MA(q) is noted as $y_t = \mu + \epsilon_t + \sum_{n = 0}^{q}\theta_n\times\epsilon_{t-n}$. Here q is the time lag that represent how many previous steps should be considered while calcualting $y_t$.

Here is the example of $MA(1)$: 

$y_t= \mu + \epsilon_t + \theta_1Ã—y_{tâˆ’1}$


```python
# Moving average model
# by setting the zeroth-order AR model with 0 and specify the order of the MA model 
model = ARMA(ar_train, order=(0, 1))
model_fit = model.fit()

print('Lag: %s' % model_fit.k_ar)
# print('Coefficients: %s' % model_fit.params)

predictions = model_fit.predict(start=len(ar_train), end=len(ar_train)+len(ar_test)-1, dynamic=False)

error = mean_squared_error(ar_test, predictions)
print('Test MSE: %.3f' % error)
# plot results
preiction  = pd.DataFrame(data = predictions)
preiction.columns = ['Prediction']
preiction.reset_index(drop=True, inplace=True)
ar_test_df['Search_AD'].reset_index(drop=True, inplace=True)
result = pd.concat([ar_test_df['Search_AD'], preiction], axis = 1, ignore_index=False)
result.plot()
```

## ARMA and ARIMA
1. ARMA: autoregression moving average model
    - Combine autoregression and moving average, denote ARMA(p, q) as $y_t = \sum_{n = 1}^{p}\phi_n \times y_{t-n} +  \epsilon_t + \sum_{n = 0}^{q}\theta_n\times\epsilon_{t-n}$
2. ARIMA: autoregression intergrated moving average model.
    - ARMA with prediction of difference between current observation and previous observation $y_{t-n}$. The model ARIMA(p, d, q) is denoted as  $z_t = y_t - y_{t-d} = \sum_{n = 1}^{p}\phi_n \times y_{t-n} +  \epsilon_t + \sum_{n = 0}^{q}\theta_n\times\epsilon_{t-n}$
    - Process of choosing the param of ARIMA - Box_Jekins method:
        - Step1. Check data and choose best model for prediction
            - Stationary checking
            - Choosing the best param base on ACF/PACF chart for AR, MA, ARMA, and ARIMA
        - Step2. Find the model that best for observed data, smallest residual
        - Step3. Diagnose Good or Bad of our model
            - Overfitting checking
            - Residual distribution checking: residual distribution should be randomness and low ACF/PACF.
        - Step4. Back to Step1 if the diagnose part is not satisified
