```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import networkx as nx

from operator import itemgetter
from numpy import genfromtxt

%matplotlib inline
```

### Load social network adjacency matrix

```python
df = pd.read_excel("./data/Alibaba-network.xlsx", index_col = 0)
df_info = df.describe().loc[['min', 'max','mean']].astype(int)
adjacency = df.values

print('Unique Value: ', df_info.loc[['max']].squeeze().unique())
```

```python
# Get networkx Graph
G = nx.from_numpy_matrix(adjacency.astype('int32'))
# G.edges(data=True)
```

### Visualize social network

```python
plt.figure(figsize=(18, 16))
nx.draw(G, node_size=20, node_color="b")
# plt.savefig('./data/Alibaba-network.png')
```

```python
# the ratio of actual edges in the network to all possible edges in the network
density = nx.density(G)
density
```

# Degree

```python
degree = nx.degree(G)
```

```python
def sort_degree(degreeView, descend=True):
    result = sorted(degreeView, key=lambda x: x[1], reverse = descend)
    return result


top_10_degree = [str(item[0]+1) for item in sort_degree(degree)[:10]]

print('Degree Max: ' + str(max(dict(degree).values())))
print('Degree Min: ' + str(min(dict(degree).values())))
print('Top 10 degree users: ' + ", ".join(top_10_degree))
```
